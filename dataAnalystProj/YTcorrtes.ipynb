```python
import pandas as pd
import seaborn as sea
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read data

df = pd.read_csv(r'F:\Marsa\Project Data\PORTO\course\dataAnalystPortoYT\dataAnalystProj\paiton\archive\movies.csv')
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
datatype = df.dtypes
print(datatype)
```

```python
df = df.astype({'budget':'int64', 'votes':'int64','gross':'int64'})
```

```python
datatype = df.dtypes
print(datatype)
```

```python
df['realyear'] = df['released'].astype(str).str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df[df['director'] == 'Christopher Nolan'][df['score'] > 8.0]
```

```python
df = df.sort_values(by=['gross'],inplace = False, ascending=False)
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.xlabel('Budget')
plt.ylabel('gross')
plt.title('Budget vs Gross Earning')
```

```python
df.head()
```

```python
#plotting regress plot

sea.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'green'})
```

```python
corr_matrix = df.corr()
print(corr_matrix)
```

```python
sea.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr_pairs = corr_matrix.unstack()
sorted_pair = corr_pairs.sort_values(ascending=False)
sorted_pair
```
